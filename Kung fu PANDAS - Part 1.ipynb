## 5. Filter Rows 

![Filter](https://j.gifs.com/76G1gG.gif "Let Go me , Let Go me")

```python

#filter rows of pandas data frame by column value
```

```python
movies=pd.read_csv('C:/Users/J.A.R.V.I.S/Desktop/pandas_data/imdb.csv')
```

```python
movies.head()
```

```python
movies.shape
```

```python
# We will examine movies that are really long.we want to filter the df such that we can see all
# the columns but only those rows where duration >=200
```

```python
type(True)
```

```python
# we are gonna create a python list of booleans that has the same exact length as the dataframe and the 
# booleans will be true if the given row is atleast 200 min and false if not
```

```python
# basically a list of booleans thats like false ,f,f, t,t,t etc
```

```python
# so we are creating a python list of booleans which tells us which rows match a given condn
```

```python
boolean=[]
for length in movies.duration:
    if length >=200:
        boolean.append(True)
    else:
        boolean.append(False)
```

```python
boolean[0:5]
```

```python
len(boolean)
```

```python
#convert boolean list to a pandas series
```

```python
is_long=pd.Series(boolean)
```

```python
is_long.head()
```

```python
movies[is_long] #Filter only the movies whose duration is greater than 200 
```

```python
type(is_long)
```

```python
                                    # or
```

```python
is_long=movies.duration >=200
is_long.head()
movies[is_long]
```

```python
                                #or
```

```python
movies[movies.duration >=200]
```

```python
# to get the genre of those >=200 duration
```

```python
movies[movies.duration >=200].genre
movies[movies.duration >=200]['genre'] #chaining is not good
#best practise is to us .loc command
movies.loc[movies.duration >=200 ,'genre']
```
